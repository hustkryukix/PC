# Code

```python
import numpy as np
import pydicom as dicom
import matplotlib.pylab as plt
import cv2
import os
import dicom2jpg as dcm2jpg
```

```python
def equalizeHist(img):
  img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
  img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
# convert the YUV image back to RGB format
  img_output = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)
  return img_output
```

```python
def draw_mask(img, mask):
    img = equalizeHist(img)
    ret, thresh = cv2.threshold(mask,0,255,cv2.THRESH_BINARY)
    thresh = np.array(thresh, dtype=np.uint8)
    edged = cv2.Canny(thresh, 30, 200)
    contours, hierarchy = cv2.findContours(edged, cv2.cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    img_out = cv2.drawContours(img, contours, -1, (255,0,0), 1)
    return img_out
```
